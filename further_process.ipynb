```python
import collections

def flatten(l):
    for el in l:
        if isinstance(el, collections.Iterable) and not isinstance(el, str):
            for sub in flatten(el):
                yield sub
        else:
            yield el
```

## Get top 50 labels

```python
from collections import Counter
```

```python
df['LABELS'] = df['LABELS'].apply(lambda x: x.split(';'))
```

```python
df['LABELS'] = df['LABELS'].apply(lambda x: list(x))
```

```python
labels = list(df['LABELS'])
```

```python
len(labels)
```

```python
len(list(flatten(labels)))
```

```python
for i, row in enumerate(labels):
    for j, code in enumerate(row):
        if code == 'nan':
            del row[j]
```

```python
len(labels)
```

```python
len(list(flatten(labels)))
```

```python
df['LABELS'] = labels
```

```python
df['LABELS'] = df['LABELS'].apply(lambda x: set(x))
labels = list(df['LABELS'])
```

```python
all_labels = []

for label_set in labels:
    for label in label_set:
        all_labels.append(label)
    
```

```python
len(all_labels)
```

```python
labels_count = Counter(all_labels)
```

```python
labels_count
```

```python
# Number of unique labels

len(labels_count)
```

```python
top_50_tuples = labels_count.most_common(50)
```

```python
top_50_tuples
```

```python
top_50 = []

for tup in top_50_tuples:
    top_50.append(tup[0])
```

```python
top_50 = set(top_50)
```

```python
df_top50 = df.copy()
df_top50['LABELS'] = df_top50['LABELS'].apply(lambda x: x.intersection(top_50))
```

```python
df_top50.head()
```

```python
len(df_top50)
```

```python
drop_indices = (df_top50[df_top50['LABELS'].map(len) == 0]).index
df_top50 = df_top50.drop(drop_indices)
```

```python
len(drop_indices)
```

```python
df_top50 = df_top50.reset_index(drop=True)
```

```python
len(df_top50)
```

```python
len(df)
```

```python
df.head()
```

## One-hot encode

```python
df_top50['LABELS'] = df_top50['LABELS'].apply(lambda x: list(x))
df['LABELS'] = df['LABELS'].apply(lambda x: list(x))
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()

df = df.join(pd.DataFrame(mlb.fit_transform(df.pop('LABELS')),
                          columns=mlb.classes_,
                          index=df.index))

df_top50 = df_top50.join(pd.DataFrame(mlb.fit_transform(df_top50.pop('LABELS')),
                                     columns=mlb.classes_,
                                     index=df_top50.index))
```

So we see that 2699 patients are dropped after removing entries which do not include a top 50 label set (3195 in total). This means that repeat patients generally have codes that belong to the common set or else we wouldn't be dropping so many patients. For example, if we lost all 3195 patients, that means these were all unique patients with a single visit (since we're dropping on a row by row basis and this would mean that this patient is no longer in the dataset). 

Anyways, this is interesting to keep note of. It seems that rare codes tend to belong to patients that visit only once.

```python
# Now we make train, val, test splits using predefined splits via CAML code repo.
```

```python
train_full_hadm_ids = pd.read_csv('data_prebert/train_full_hadm_ids.csv', header=None)
dev_full_hadm_ids = pd.read_csv('data_prebert/dev_full_hadm_ids.csv', header=None)
test_full_hadm_ids = pd.read_csv('data_prebert/test_full_hadm_ids.csv', header=None)
```

```python
train_50_hadm_ids = pd.read_csv('data_prebert/train_50_hadm_ids.csv', header=None)
dev_50_hadm_ids = pd.read_csv('data_prebert/dev_50_hadm_ids.csv', header=None)
test_50_hadm_ids = pd.read_csv('data_prebert/test_50_hadm_ids.csv', header=None)
```

```python
frames = [train_50_hadm_ids, dev_50_hadm_ids, test_50_hadm_ids]
full_50_hadm_ids = pd.concat(frames)
```

```python
len(full_50_hadm_ids)
```

```python
df_train_full = df[df['HADM_ID'].isin(list(flatten(train_full_hadm_ids.values)))]
df_dev_full = df[df['HADM_ID'].isin(list(flatten(dev_full_hadm_ids.values)))]
df_test_full = df[df['HADM_ID'].isin(list(flatten(test_full_hadm_ids.values)))]
```

```python
len(df_train_full), len(df_dev_full), len(df_test_full)
```

```python
df_train_50 = df_top50[df_top50['HADM_ID'].isin(list(flatten(train_50_hadm_ids.values)))]
df_dev_50 = df_top50[df_top50['HADM_ID'].isin(list(flatten(dev_50_hadm_ids.values)))]
df_test_50 = df_top50[df_top50['HADM_ID'].isin(list(flatten(test_50_hadm_ids.values)))]
```

```python
len(df_train_50), len(df_dev_50), len(df_test_50)
```

```python
df_train_full = df_train_full.drop(columns=['SUBJECT_ID', 'HADM_ID'])
```

```python
df_dev_full = df_dev_full.drop(columns=['SUBJECT_ID', 'HADM_ID'])
df_test_full = df_test_full.drop(columns=['SUBJECT_ID', 'HADM_ID'])
```

```python
df_train_50 = df_train_50.drop(columns=['SUBJECT_ID', 'HADM_ID'])
df_dev_50 = df_dev_50.drop(columns=['SUBJECT_ID', 'HADM_ID'])
df_test_50 = df_test_50.drop(columns=['SUBJECT_ID', 'HADM_ID'])
```

```python
# Our splits are ready. Save to disk

df_train_full.to_csv('data_prebert/train_full.csv', index=False)
df_dev_full.to_csv('data_prebert/val_full.csv', index=False)
df_test_full.to_csv('data_prebert/test_full.csv', index=False)

df_train_50.to_csv('data_prebert/train_50.csv', index=False)
df_dev_50.to_csv('data_prebert/val_50.csv', index=False)
df_test_50.to_csv('data_prebert/test_50.csv', index=False)
```
